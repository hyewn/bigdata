```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rc('font', family='NanumGothic') # For Windows
# font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf' # For Linux/Mac, 폰트 경로를 자신의 시스템에 맞게 수정
# fontprop = fm.FontProperties(fname=font_path, size=18)

# 데이터 생성
data = {
    '한강공원': ['여의도한강공원', '이촌한강공원', '망원한강공원', '양화한강공원', '뚝섬한강공원', '잠실한강공원', '서울숲', '잠원한강공원(압구정)'],
    '지하철 노선 수': [2, 3, 2, 2, 2, 3, 2, 1],
    '버스 노선 수': [24, 16, 2, 5, 10, 27, 17, 25]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 데이터프레임 출력 (확인용)
print(df)

# 막대 그래프 생성
fig, ax = plt.subplots(figsize=(12, 8))

# 막대 그래프 너비 설정
width = 0.35  # 막대 너비
x = range(len(df['한강공원']))

# 지하철 노선 수에 대한 막대
ax.bar(x, df['지하철 노선 수'], width, label='지하철 노선 수')

# 버스 노선 수에 대한 막대
ax.bar([p + width for p in x], df['버스 노선 수'], width, label='버스 노선 수')

# 한강공원 이름을 x축에 추가
ax.set_xticks([p + width/2 for p in x])
ax.set_xticklabels(df['한강공원'], rotation=45)

# 그래프 제목과 축 이름 설정
ax.set_title('한강공원으로의 접근성 데이터')
ax.set_xlabel('한강공원')
ax.set_ylabel('노선 수')

# 범례 추가
ax.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```
